#### **train Îç∞Ïù¥ÌÑ∞ÏôÄ validation Îç∞Ïù¥ÌÑ∞ ÎÇòÎàÑÍ∏∞**

```python
from sklearn.model_selection import train_test_split
import pandas as pd


df=pd.read_csv('/content/drive/MyDrive/EMRM_Detection_data/train.txt',header=None,names=['Image_name'])
df.head()
```

```python
train_df,val_df=train_test_split(df,test_size=0.2,random_state=2021)
print(train_df.shape,val_df.shape)
```

```python
!mkdir /content/drive/MyDrive/EMRM_Detection_data/images/train
!mkdir /content/drive/MyDrive/EMRM_Detection_data/images/val
!mkdir /content/drive/MyDrive/EMRM_Detection_data/labels/train
!mkdir /content/drive/MyDrive/EMRM_Detection_data/labels/val
```

```python
import shutil

images_path='/content/drive/MyDrive/EMRM_Detection_data/images/'
labels_path='/content/drive/MyDrive/EMRM_Detection_data/labels/'

def move_data(df,datatype):
  for index,row in df.iterrows():
    image_path=images_path+row['Image_name']+'.jpg'
    label_path=labels_path+row['Image_name']+'.txt'

    tgt_image_path=images_path+datatype+'/'+row['Image_name']+'.jpg'
    tgt_labels_path=labels_path+datatype+'/'+row['Image_name']+'.txt'
    shutil.copy(image_path,tgt_image_path)
    shutil.copy(label_path,tgt_labels_path)

move_data(train_df,'train')
move_data(val_df,'val')
```
